```python
import pickle
import pandas as pd
import numpy as np
import time

from cluster_report import generate_report_for_cluster, group_cluster_indexes

pd.options.display.max_columns = 200
pd.options.display.max_rows = 100
pd.options.display.max_colwidth = 1000
```

# PIE

```python
with open("../data/dump_pie.pickle", "rb") as f:
    dump_wl = pickle.load(f)
    
texts = np.array(dump_wl['texts'])
clusters = np.array(dump_wl['clusters'])
vectors = np.array(dump_wl['vectors'])
cluster_dict = group_cluster_indexes(clusters)

for cluster_id in range(1, 51):
    generate_report_for_cluster(cluster_id, texts, clusters, vectors, cluster_dict, k=10)
    time.sleep(0.5)
```
